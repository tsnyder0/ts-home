```python
import numpy as np
import matplotlib.pyplot as plt
import math
import re
```

## For Start Time

```python
times=[]

filename='10-14-22-STARLINK-passes-python.csv'
datafile=open(filename,'r')
for line in datafile:
    if (line[0]=='#'):
        print(' ')
    else:
        line1=re.split(',|:',line)
        line1[1]=re.sub('13-Oct','2022-10-13',line1[1])
        line1[1]=re.sub('14-Oct','2022-10-14',line1[1])
        line1[1]=re.sub('15-Oct','2022-10-15',line1[1])
        del line1[2:4],line1[5:23],line1[0]
        if (line1[1]=='Start'):
            print(' ')
        else:
            #print(line1)
            line2 = ':'.join(line1[1:4])
            line3 = line1[0] + "T" + line2
            #print(line3)
            times.append(line3)


time = np.array(times)
print(time)
```

```python
from astropy.time import Time, TimeDelta

t = Time(times, format='isot', scale='utc')

td = TimeDelta(7200, format='sec')
```

```python
t = t + td

print(t)
```

```python
for t in t:
    print(t.jd)
```
